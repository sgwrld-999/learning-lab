```python
# standards imports
import os
from typing import TypedDict
from dotenv import load_dotenv

# third-party imports
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
load_dotenv()
```

```python
gemini_model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"#,
    # google_api_key=api_key,
)
```

```python
# define state
class LLMState(TypedDict):
    
    question : str
    answer   : str
```

```python
# define question state
def llm_ques(state: LLMState) -> LLMState:
    
    # extract the question
    question = state["question"]
    
    prompt = f'Your task is to answer the following {question} in a simple, neat and clean manner. No need to over-explain anything just answer in the simple words'
    
    # ask the question to the LLM
    answer = gemini_model.invoke(prompt).content
    
    # updating the state
    state["answer"] = answer
    
    return state
```

```python
# define graphs
graph = StateGraph(LLMState)

# add nodes
graph.add_node('llm_ques', llm_ques)

# add edges 
graph.add_edge(START, "llm_ques")
graph.add_edge("llm_ques", END)

# compile
workflow = graph.compile()
```

```python
# execute the graph
initial_state = {'question' : 'What is the your name?'}

final_state = workflow.invoke(initial_state)

print(final_state['answer'])
```
