### Decorator

```python
def decorator(func): #Can recieve arguments or takes as input
    def wrapper(*args, **kwargs): #will be created at the run time
        print("Before Print")
        result = func(*args, **kwargs)
        print("After print")
        return result
    return wrapper #returning the modified function
```

```python
@decorator # -> Annotation for the decorator
def greet(name):
    print(f"Hello, {name}")
```

```python
greet("If you smell, what are the Rock? is cooking!!!")
```

```python
def countdown_generator(num: int):
    while num >= 0:
        yield num
        num -= 1

def even_number(start: int, end: int):
    for num in range(start, end + 1):
        if num % 2 == 0:
            yield num

def repeated_character(ipStr: str):
    for ch in ipStr:
        yield ch * 2
```

```python
def calculate(a: float, b: float, op: str) -> float:
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    else:
        raise ValueError("Invalid operation")

# Example
calculate(10.5, 2.0, '*')
```

```python
calculate("2","1","*")
```
